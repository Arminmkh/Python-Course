# Some built-in functions:
str() is a built-in function that changes data type to string.

```python
n = 5
s = str(n)
type(s)
```

print() is a built-in function that shows a value in the output.

```python
#Guess the output before execution.
print(n*2)
print(s*2)
```

```python
print(s*2 + n*2)
```

len() is a built-in function that gives the length of an object.

```python
print(len(s*2))
```

```python
variables = [n*2, s*2] #What is type of this variable?
len(variables)
```

```python
print(len(n*2))
```

len()'s input can be:
sequence (string, bytes, tuple, list, range)
OR collection (dictionary, set, frozen set)

The number of Python functions is unlimited since people use def to define their own functions.

# Function in function

```python
def Devide(n):
    return n/1024

def Bytes_to_Megabytes(Devide, n):
    return Devide(Devide(n))
```

```python
volume = 10**7 #Bytes
print(volume,"Bytes is equal to")
Bytes_to_Megabytes(Devide, volume)
```
